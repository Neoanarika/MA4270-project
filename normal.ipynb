```
import numpy as np
from scipy.stats import gamma, norm, expon
import matplotlib.pyplot as plt
```

```
plt.style.use('ggplot')
```

```
a = 25
normRV = norm(0, 1000)
normRV2 = norm(100, 1000)
```

```
x = np.linspace(normRV.ppf(0.01),
                normRV.ppf(0.99), 100000)
y = np.linspace(normRV2.ppf(0.01),
                normRV2.ppf(0.99), 100000)
```

```
fig, ax = plt.subplots(1, 1)
ax.plot(x, normRV.pdf(x),
       'r-', lw=1, alpha=0.6, label='N(0, 1000)')
ax.plot(y, normRV2.pdf(y),
       'b-', lw=1, alpha=0.6, label='N(100, 1000)')
ax.legend(loc='best', frameon=False)
plt.show()
```

```
normRV4 = norm(0, 0.1)
normRV3 = norm(0.1, 0.1)

a = np.linspace(normRV4.ppf(0.01),
                normRV4.ppf(0.99), 100000)
b = np.linspace(normRV3.ppf(0.01),
                normRV3.ppf(0.99), 100000)
```

```
fig, ax = plt.subplots(1, 1)
ax.plot(a, normRV4.pdf(a),
       'r-', lw=1, alpha=0.6, label='N(0, 0.1)')
ax.plot(b, normRV3.pdf(b),
       'b-', lw=1, alpha=0.6, label='N(0, 0.5)')
ax.legend(loc='best', frameon=False)
plt.show()
```

```
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(13, 5))
fig.suptitle("Comparsion between 2 pairs of normal distribution")
axes[0].plot(x, normRV.pdf(x),
       'r-', lw=1, alpha=0.6, label='N(0, 1000)')
axes[0].plot(y, normRV2.pdf(y),
       'b-', lw=1, alpha=0.6, label='N(100, 1000)')
axes[0].set_title('N(0, 1000) vs N(100, 1000)')
axes[0].legend(loc='best', frameon=False)
axes[1].plot(a, normRV4.pdf(a),
       'r-', lw=1, alpha=0.6, label='N(0, 0.1)')
axes[1].plot(b, normRV3.pdf(b),
       'b-', lw=1, alpha=0.6, label='N(0.1, 0.1)')
axes[1].set_title('N(0, 0.1) vs N(0.1, 0.1)')
axes[1].legend(loc='best', frameon=False)
```
